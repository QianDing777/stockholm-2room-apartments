```python
# Group the data by area and calculate the desired statistics
grouped = df.groupby('area').agg(['mean', 'std', 'min', 'max', lambda x: x.quantile(0.25), lambda x: x.quantile(0.5), lambda x: x.quantile(0.75)])

# Rename the columns to be more descriptive
grouped.columns = [f"{col}_{agg}" if agg != '' else col for col, agg in grouped.columns]

# Print the grouped data
print(grouped)
```
