## 2. Data Collection

The columns in our dataset scraped from Hemnet.se can be described as follows:

 - __area__: The area in Stockholm where the property is located.
 - __adress__: The street address of the property.
 - __price__: The selling price of the property in SEK.
 - __price_per_sqm__: The selling price per square meter of the property.
 - __monthly_fee__: The monthly fee associated with owning the property, such as maintenance fees or homeowner association fees.
 - __size_in_sqm__: The total size of the property in square meters.
 - __date_sold__: The date on which the property was sold.
 
The source of the data is _Hemnet.se_, a popular Swedish website for buying and selling properties. The website provides listings of properties for sale, along with various details such as the location, price, size, and fees.

The dataset scraped contains over 2000 records and 7 columns.
